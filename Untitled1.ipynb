```python
from datetime import date
from nsepy.history import get_price_list
import pandas as pd
import numpy as np
i = 11
prices1 = pd.DataFrame(get_price_list(dt=date(2021,5,i)))
prices1 = prices1.rename(columns={"SERIES":("SERIES"+str(i)),"OPEN":("OPEN"+str(i)),"HIGH":("HIGH"+str(i)),"LOW":("LOW"+str(i)),"CLOSE":("CLOSE"+str(i)),"LAST":("LAST"+str(i)),"PREVCLOSE":("PREVCLOSE"+str(i)),"TOTTRDQTY":("TOTTRDQTY"+str(i)),"TOTTRDVAL":("TOTTRDVAL"+str(i)),"TIMESTAMP":("TIMESTAMP"+str(i)),"TOTALTRADES":("TOTALTRADES"+str(i)),"ISIN":("ISIN"+str(i))})
x = {12,14,17,18,19,20,21,24,25}
for i in x:
    
    prices = pd.DataFrame(get_price_list(dt=date(2021,5,i)))
    prices = prices.rename(columns={"SERIES":("SERIES"+str(i)),"OPEN":("OPEN"+str(i)),"HIGH":("HIGH"+str(i)),"LOW":("LOW"+str(i)),"CLOSE":("CLOSE"+str(i)),"LAST":("LAST"+str(i)),"PREVCLOSE":("PREVCLOSE"+str(i)),"TOTTRDQTY":("TOTTRDQTY"+str(i)),"TOTTRDVAL":("TOTTRDVAL"+str(i)),"TIMESTAMP":("TIMESTAMP"+str(i)),"TOTALTRADES":("TOTALTRADES"+str(i)),"ISIN":("ISIN"+str(i))})
    # s = str(i)+'.csv'
    # prices.to_csv(s)
    prices1 = prices1.merge(prices, left_on='SYMBOL', right_on='SYMBOL', how='outer',)
prices1 = prices1.dropna()
prices1 = prices1.sort_values('TOTTRDVAL25',ascending=False)
prices2 = prices1.iloc[:,[0,12,2,3,4,5,14,15,16,17,26,27,28,29,38,39,40,41,50,51,52,53,62,63,64,65,74,75,76,77,86,87,88,89,98,99,100,101,110,111,112,113]]
column1 = ['SYMBOL','ISIN','O9','H9','L9','C9','O8','H8','L8','C8','O7','H7','L7','C7','O6','H6','L6','C6','O5','H5','L5','C5','O4','H4','L4','C4','O3','H3','L3','C3','O2','H2','L2','C2','O1','H1','L1','C1','O','H','L','C']
prices2.columns = column1
prices2.insert(42,"AVGH10", (prices2["H9"]+prices2["H8"]+prices2["H7"]+prices2["H6"]+prices2["H5"]+prices2["H4"]+prices2["H3"]+prices2["H2"]+prices2["H1"]+prices2["H"])/10)
prices2.insert(43,"AVGL10", (prices2["L9"]+prices2["L8"]+prices2["L7"]+prices2["L6"]+prices2["L5"]+prices2["L4"]+prices2["L3"]+prices2["L2"]+prices2["L1"]+prices2["L"])/10)
prices2.insert(44,"MINL10",prices2.min(axis = 1) )
prices2.insert(45,"GREEN",(prices2["C"]>=prices2["O"]))
prices2.insert(46,"ENGULFING",(prices2["O1"]>prices2["C1"]) & (10*(prices2["C"]-prices2["O"])>= 7*(prices2["H"]-prices2["L"])) & (prices2["C"]>prices2["O1"]) & (prices2["C1"]>prices2["O"]) & (10*(prices2["H"]-prices2["L"])>=12*(prices2["AVGH10"]-prices2["AVGL10"])))
# prices3 = prices2.loc[prices2['ENGULFING'] == 'TRUE' ]
prices2.insert(47,"DojiDragonfly",((50*np.absolute(prices2["O"]-prices2["C"]))<=(prices2["H"]-prices2["L"])) & ((prices2["H"]-prices2["L"]) >= (prices2["AVGH10"]-prices2["AVGL10"])) & prices2["L"] == prices2["MINL10"]   )
prices1.to_csv('final.csv')
prices2.to_csv('final1.csv')
# prices3.to_csv('ENGULFING.csv')
```
